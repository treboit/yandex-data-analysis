### Предобработка данных<a id="step1_3"></a>
Переименовываю столбцы:

```python
order.columns = ['buy_time', 'revenue', 'uid']
visit.columns = ['device', 'end_time', 'source_id', 'start_time', 'uid']
cost.columns = ['source_id', 'date', 'costs']
```

Выручка и расходы отражены в тысячах (предположительно, рублей). Перевожу в рубли, чтобы привести к целочисленному типу данных.

```python
order['revenue'] = (order['revenue'] * 1000).astype('int')
cost['costs'] = (cost['costs'] * 1000).astype('int')
```

Привожу данные времени и даты к соответствующему типу

```python
order['buy_time'] = pd.to_datetime(order['buy_time'])
visit['end_time'] = pd.to_datetime(visit['end_time'])
visit['start_time'] = pd.to_datetime(visit['start_time'])
cost['date'] = pd.to_datetime(cost['date'])
```

В данных по посещениям добавляю информацию о продолжительности сессии и ее дате (по дате начала):

```python
visit['ses_length'] = visit['end_time'] - visit['start_time']
visit['ses_date'] = visit['start_time'].dt.date
visit['ses_week'] = visit['start_time'].dt.week
visit['ses_month'] = visit['start_time'].astype('datetime64[M]')
```

В данных по продажам и расходам добавляю информацию о месяце:

```python
order['buy_month'] = order['buy_time'].astype('datetime64[M]')
cost['cost_month'] = cost['date'].astype('datetime64[M]')
```

Данные предобработаны, пропусков нет, типы данных корректны.

**Сколько людей пользуются в день, неделю, месяц?**<a id="step2_1_1"></a>

  
Считаю количество уникальных пользователей для каждого дня, недели, месяца и года в целом:

```python
dau = visit.groupby('ses_date')['uid'].nunique()
wau = visit.groupby('ses_week')['uid'].nunique()
mau = visit.groupby('ses_month')['uid'].nunique()
for data in [dau, wau, mau]:
    print(data.mean().round())
print(visit['uid'].nunique())
```

```python
(visit['uid'].nunique() / dau.mean() / 365).round(2)
```

Большая часть пользователей (69%) заходит на Яндекс.Афишу в какой-то один день и не возвращается в течение года, но остальная треть пользуется сервисом чуть более регулярно. Это ожидаемо с учетом того, что люди нечасто ходят на крупные события, билеты на которые продаются на Афише. Также на количество уникальных пользователей влияет невозможность сопоставить пользователя, если он зашел с другого устройства без своего аккаунта.

**Как часто люди возвращаются?** <a id="step2_1_4"></a>

 
Строю распределение user retention rate по когортам, сформированным по месяцам от момента первой активности:

```python
visit['c_lifetime'] = ((visit['ses_month'] - visit['fam']) / np.timedelta64(1,'M')).round().astype(int)
cohorts = visit.groupby(['fam','c_lifetime']).agg({'uid':'nunique'}).reset_index()
inital_users_count = cohorts[cohorts['c_lifetime'] == 0][['fam','uid']]
inital_users_count = inital_users_count.rename(columns={'uid':'cohort_users'})
cohorts = cohorts.merge(inital_users_count,on='fam')
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']
retention_pivot = cohorts.pivot_table(index='fam',columns='c_lifetime',values='retention',aggfunc='sum')
retention_pivot.index = retention_pivot.index.strftime('%Y-%m')
```

```python
plt.figure(figsize=(10, 5))
plt.title('User Retention')
ax = sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmin = 0, vmax = 0.1)
ax.set_xlabel('Возраст когорты')
ax.set_ylabel('Когорта пользователей по первой активности')
```

График по когортам подтверждает вывод, сделанный раньше: большая часть пользователей использует Афишу только один раз, помесячная метрика retention rate низкая. 

```python
cohorts.groupby('c_lifetime')['retention'].mean()[1:].plot(kind='bar').set(
    xlabel='Время жизни когорты', ylabel='Retention rate')
```

Можно было бы ожидать, что средний retention более ранних когорт ниже последних, так как пользователи со временем уходят, однако фактически на графике ниже видна обратная ситуация: за небольшими исключениями retention снижается со временем. Это явный негативный показатель продукта.

```python
cohorts_retention = cohorts[cohorts['c_lifetime'] != 0].groupby('fam')['retention'].mean()
cohorts_retention.index = cohorts_retention.index.strftime('%Y-%m')
cohorts_retention.plot(kind='bar').set(xlabel='Когорта', ylabel='Retention rate')
```

Основываясь на распределении по времени жизни когорты (за исключением нулевого месяца), видно, что в течение оставшего срока жизни возвращается примерно одинаковая доля пользователей (4,5% - 6%). Это могло бы говорить о том, что оттока пользователей не происходит, просто потребность в сервисах Афиши очень редкая. Однако, если вернуться к графику по отдельным когортам выше, видно, что retention выше 4% сохраняется на всех периодах только у двух первых когорт. Retention следующих когорт быстро снижается до 2%-3%, поэтому вероятно, что в последующих периодах, основанных на данных новых когорт, retention будет ниже в 2 раза. Динамика негативная, но без дальнейшего анализа нельзя говорить о снижении качества продукта:  
- во-первых, нужно понимать, какой retention является нормальным для продукта, возможно, снижение вызвано ненормально высокими показателями первых когорт;
- во-вторых, важно то, что пользователи заходят на Афишу, чтобы совершить покупку. Чтобы корректно оценивать rentention, нужно проанализировать продажи.  

### Продажи<a id="step2_2"></a>

**Когда люди начинают покупать?**<a id="step2_2_1"></a>

Для начала следует учесть долю посетителей, которые оформляют заказ:

```python
print('Доля посетителей, оформивших заказ: {:.0%}'.format(order['uid'].nunique() / visit['uid'].nunique()))
```

Теперь, не учитывая посетителей, не оформивших заказ, составляю таблицу со временем первого посещения и первой покупки для каждого пользователя, а также времени между ними:

```python
first_order = visit.groupby('uid', as_index = False)['start_time'].min()\
    .merge(order.groupby('uid', as_index = False)['buy_time'].min(), how = 'inner', on = 'uid')
first_order.columns = ['uid', 'first_session', 'first_order']
first_order['start2order'] = first_order['first_order'] - first_order['first_session']
first_order[:1]
```

```python
print('Медианное время от первого посещения до заказа: ', first_order['start2order'].median(), 
      ', среднее время: ', first_order['start2order'].mean())
```

Половина пользователей делает заказ в течение первых 16 минут, другая может сделать его в следующие сессии в другие дни. Строю гистограмму для покупок тремя частями:  
- в течение первого часа после первого посещения,
- в течение первых суток, начиная со второго часа,
- в оставшееся время, начиная со второго дня после первого посещения:

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (18,5))
sns.distplot((first_order['start2order'] / np.timedelta64(1,'m')).where(lambda x : x<60).dropna().round(), 
    ax = axes[0], bins = 50, color = plt.cm.tab10(0)).set(xlabel='Минуты с начала посещения', ylabel='Доля заказов')
sns.distplot((first_order['start2order'] / np.timedelta64(1,'h')).where(lambda x : x<=24).where(lambda x : x>1).dropna().round(), 
    ax = axes[1], bins = 50, color = plt.cm.tab10(1)).set(xlabel='Часы в первые сутки с начала посещения', ylabel='Доля заказов')
sns.distplot((first_order['start2order'] / np.timedelta64(1,'D')).where(lambda x : x>=1).dropna().round(), 
    ax = axes[2], bins = 50, color = plt.cm.tab10(2)).set(xlabel='Дни со второго дня после начала посещения', ylabel='Доля заказов')
```

По графикам можно сделать выводы о том, что:  
- значительная часть пользователей совершает заказ в первую минуту после захода на Афишу; это те пользователи, которые зашли на Афишу только оформить покупку билетов на мероприятие, которое они уже выбрали вне Афиши. Эти же пользователи формируют медианное время сессии на сайте в 5 минут;
- далее, за исключением отдельных пиков, распределение убывающее, что было понятно из медианного времени от захода до заказа в 16 минут;
- на дневном графике заметно увеличение заказов в районе 20-24 часов: это пользователи, не оформившие заказ накануне, примерно в это же время на следующий день возвращаются к сервису, чтобы оформить заказ;
- график за все время нормально убывает, так как новые пользователи приходят постоянно, а данные по заказам есть только за один год; здесь применим анализ по когортам.

```python
first_order['fom'] = first_order['first_order'].astype('datetime64[M]') #first order month
first_order['fam'] = first_order['first_session'].astype('datetime64[M]') #first activity month
first_order['c_lifetime'] = ((first_order['fom'] - first_order['fam']) / np.timedelta64(1,'M')).round().astype(int)
```

```python
cohorts = first_order.groupby(['fam','c_lifetime']).agg({'uid':'nunique'}).reset_index()
cohort_users_count = cohorts.groupby('fam')['fam','uid'].sum()
cohort_users_count = cohort_users_count.rename(columns={'uid':'cohort_users'})
cohorts = cohorts.merge(cohort_users_count,on='fam')
cohorts['order_share'] = cohorts['uid']/cohorts['cohort_users']
order_pivot = cohorts.pivot_table(index='fam',columns='c_lifetime',values='order_share',aggfunc='sum')
order_pivot.index = order_pivot.index.strftime('%Y-%m')
```

```python
plt.figure(figsize=(10, 5))
plt.title('Доля первых покупок от даты первой сессии')
ax = sns.heatmap(order_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmin = 0, vmax = 0.1)
ax.set_xlabel('Возраст когорты')
ax.set_ylabel('Когорта пользователей по первой активности')
```

Увеличение доли покупок в нулевой месяц каждой следующей когорты происходит из-за того, что в график не включены пользователи, еще не совершившие свой заказ. Поэтому нужно анализировать долю покупок в каждый месяц жизни (возраст) когорт. Здесь видно, что доля покупок в последующие месяцы постепенно снижается: у первых когорт значения для каждого соответсвующего месяца жизненного цикла в среднем выше, чем у следующих когорт. Это говорит о том, что все большее количество пользователей совершают заказ в первый месяц посещения Афиши, а не спустя некоторое время, что можно считать положительным изменением продукта. Также это объясняет снижение retention: вероятно, в последующие месяцы на Афишу не возвращаются пользователи, совершившие заказ после первого посещения. Таким образом, косвенно можно сделать вывод, что пользователи сразу находят на сервисе то, за чем они пришли. 

**Сколько раз покупают за период?**<a id="step2_2_2"></a>

Как уже было показано выше, большинство покупок происходит в первый месяц активности, далее они распределены примерно равномерно в оставшиеся месяцы. Я группирую данные по заказам по месяцам первого заказа и считаю показатель количества заказов на одного пользователя за каждый месяц с момента первой покупки (не считая сам первый заказ, потому что с него ведется отсчет):

**Сколько денег приносят? (LTV)**<a id="step2_2_4"></a>

Считаю показатели LTV, а также CAC и ROMI, которые будут использованы далее:

```python
cohorts = order.groupby(['fom','buy_month'], as_index = False)['revenue'].sum()
cohorts = cohorts.merge(new_order[['fom', 'uid']], on = 'fom')
cohorts['age'] = (cohorts['buy_month'] - cohorts['fom']) / np.timedelta64(1, 'M')
cohorts['age'] = cohorts['age'].round().astype(int)
cohorts['ltv'] = cohorts['revenue'] / cohorts['uid'] #ltv = выручка в пересчете на пользователя
cohorts = pd.merge(cohorts, 
                   cost.groupby('cost_month', as_index = False)['costs'].sum(), 
                   left_on='fom', right_on='cost_month', how = 'left')
cohorts['cac'] = cohorts['costs'] / cohorts['uid'] #cac = расходы в пересчете на пользователя
cohorts['romi'] = cohorts['ltv'] / cohorts['cac']
cohorts.head()
```

Накопительный LTV по когортам: 

```python
ltv = cohorts.pivot_table(index='fom', columns='age', values='ltv', aggfunc='mean').round()
ltv.cumsum(axis=1).fillna('')
```

Дополнительных комментариев не нужно, так как эта таблица, с учетом того, что количество покупок на пользователя стремится к единице, по сути аналогична накопительным данным по среднему чеку. Данные по LTV будут проанализированы применительно к ROMI.

В данном случае прирост выручки явно незначительный, эффект от маркетинговых расходов явно краткосрочный, а LTV клиента в основном сосредоточена в первой покупке. Значит, или валовая прибыль не является целью Афиши на данном этапе, или маркетинговая стратегия должна быть изменена, чтобы достичь окупаемости. Для того, чтобы судить об эффективности источников привлечения, недостаточно данных: например, для Яндекс.Афиши главным источником привлечения, вероятно, является сам Яндекс, поэтому эти расходы достаточно условны и опредяются трансфертными ценами между подразделениями компании, а канал по сути безальтернативен; а источники, которые хотя бы окупились на некоторых когортах, могут быть не масштабируемыми. Однако, основываясь на имеющихся данных, можно говорить, что источник №3 не окупается вообще, источник №2 с похожей высокой стоимостью привлечения клиента имеет больше шансов окупиться, так же, как и более дешевые источники №№ 1, 4 и 10. При этом важной целью является поиск маркетинга, который позволит привлекать постоянных клиентов и на длительный срок.   
В целом, можно остановиться на выводе о том, что на текущий момент маркетинг не окупается практически на всех источниках и периодах, однако нет данных для принятия решения о необходимости исправления ситуации.
#### Продукт
Retention достаточно низкий и в течение года снижается с 6% до 3%. Нужно оценить источники привлечения и другие метрики счастья пользователей первых когорт и выяснить причину снижения, так как сохранение пользователей является одним из основных факторов окупаемости продукта.  
В течение анализируемого периода доля пользователей, совершающих покупку в следующие за первым посещением периоды, постепенно сокращается, то есть пользователи стали чаще совершать покупку после первого посещения сервиса. Это может быть как результатом улучшения качества продукта, однако также возможно, что часть пользователей просто не возвращается на сервис за покупкой, если не совершила ее после первого посещения.  
Доля пользователей с мобильных платформ достаточно низкая для 2017-18 гг.: 27%, при этом аудитории мобильных устройств и ПК не отличаются. Нужно оценить причины этого, вероятно, мобильная аудитория недостаточно охвачена.
